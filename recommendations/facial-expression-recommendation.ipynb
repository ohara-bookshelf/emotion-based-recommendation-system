```python
import pandas as pd
import numpy as np
```

```python
# Read the csv file
survey_csv = pd.read_csv('../survey/survey.csv',on_bad_lines='skip', sep=',', encoding='latin-1')
```

```python
survey_csv.head()
```

```python
# Create a new dataframe without the first 3 columns
survey_df = survey_csv.iloc[:,3:]
```

```python
survey_df.head()
```

```python


# Get the list of emotions (excluding the first column)
emotions = survey_df.columns[1:] 
```

```python

emotions
```

```python

# Step 1: Calculate the Genre Weights for Each Column
genre_weights = {}

for emotion in emotions:
    genres = survey_df[emotion].str.split(';').explode()
    genre_counts = genres.value_counts()
    genre_weights[emotion] = genre_counts / len(survey_df)
```

```python
# Step 2: Calculate each Genre Weights for Each Column
# (Normalization is already done in the previous step)
for emotion in emotions:
    genre_weight = genre_weights[emotion]
    for genre in genre_weight.index:
        print(f"{emotion} {genre}: {genre_weight[genre]:.2f}")

```

```python

# Put those values on a table for better visualization
genre_weights_df = pd.DataFrame(genre_weights)
```

```python
genre_weights_df
```

```python
# If values are NaN, it means that the emotion is not related to that genre
# PUT IT AS ZERO INSTEAD OF NaN
genre_weights_df = genre_weights_df.fillna(0)
```

```python
# A function to return the the top genres for a given emotion
def get_top_genres(emotion, top_genres_count):
    return genre_weights_df[emotion].sort_values(ascending=False).head(top_genres_count).index.tolist()
```

```python
# Get the books dataframe
books_df = pd.read_csv('../recommendations/dataset/bx_10370_books.csv',on_bad_lines='skip', sep=',', encoding='latin-1')
```

```python
books_df.head()
```

```python
# If books genres does not have one of the survey genres, remove it to the dataframe
survey_genres = list(survey_genres)

# Create a mask to mark rows that should be removed
mask = books_df['Genres'].apply(lambda genres: not any(genre in genres for genre in survey_genres))

# Remove rows that don't have at least one survey genre
books_df = books_df[~mask]
```

```python
books_df.shape
```

```python
books_df.head()
```

```python
# Drop the books that do not have any genre
books_df = books_df.dropna(subset=['Genres'])
```

```python
#Create a new dataset with ISBN, Book-Id and Genres
books_genres_df = books_df[['ISBN', 'Book-Id', 'Genres']].copy()
```

```python
books_genres_df.head()
```

```python
pickle.dump(genre_weights_df, open("../app/model/books_genres_df-1.0.0.pkl", "wb"))
```

```python
# Save this dataframe as a pickle file
#books_genres_df.to_pickle('../app/model/books_genres_df-1.0.0.pkl')
```

```python
def recommend(emotion, books_count):
    # Get the genre weights for the given emotion
    genre_weights = genre_weights_df.loc[:, emotion]
    
    # Normalize the genre weights to sum up to 1
    genre_weights /= genre_weights.sum()
    
    # Sort the genre weights in descending order
    genre_weights = genre_weights.sort_values(ascending=False)
    
    # Select the top genres based on the weights
    top_genres = genre_weights.index.tolist()
    
    # Filter books dataframe based on top genres
    filtered_books = books_genres_df[books_genres_df['Genres'].apply(lambda genres: any(genre in genres for genre in top_genres))]
    
    # Get the desired number of books
    recommended_books = filtered_books['ISBN'].tolist()[:books_count]
    
    return recommended_books
```

```python
emotion = 'happy'  # Specify the emotion
books_count = 20  # Specify the number of books to recommend

recommended_books = recommend(emotion, books_count)
```
